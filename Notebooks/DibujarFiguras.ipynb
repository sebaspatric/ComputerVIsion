```python
# prompt: borrar uno de los circulos

imagen = np.zeros(shape=(500,500,3), dtype=np.int16)
imagen.shape
plt.imshow(imagen)
cv2.rectangle(imagen, pt1=(20,20), pt2=(100,100), color=(255,0,0), thickness=10)
plt.imshow(imagen)
cv2.rectangle(imagen, pt1=(20,20), pt2=(100,100), color=(0,255,0), thickness=10)
plt.imshow(imagen)
cv2.rectangle(imagen, pt1=(20,20), pt2=(100,100), color=(255,120,0), thickness=10)
plt.imshow(imagen)
cv2.circle(imagen, center=(250,250), radius=100, color=(0,255,0), thickness=10)
plt.imshow(imagen)

# To delete the circle, simply draw a rectangle over it with the same color as the background.
cv2.rectangle(imagen, pt1=(200,200), pt2=(300,300), color=(0,0,0), thickness=-1)
plt.imshow(imagen)
```

```python
cv2.line(imagen, pt1=(0,400),pt2=(500,400), color=(0,0,255), thickness=10)
plt.imshow(imagen)
```
